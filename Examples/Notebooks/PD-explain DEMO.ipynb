```python
spotify_all = pd.read_csv(r'C:\spotify_all.csv', encoding = 'latin-1')
```

```python
popular = spotify_all[spotify_all.popularity > 70]
```

```python
popular.explain()
```

```python
popular.explain(top_k = 3, figs_in_row = 3)
```

```python
popular.explain(top_k = 3, figs_in_row = 3, corr_TH = 0.5)
```

```python
popular.present_deleted_correlated()
```

```python
newpopular = popular[popular.year >= 1990]
by_decades = newpopular.groupby("decade").mean()
```

```python
by_decades.explain()
```

```python
by_decades.explain(top_k = 3, figs_in_row = 3)
```
